```python
import pandas as pd
import json

json_file = "response.json"
```

```python
with open(json_file, 'r') as j:
    json_dict = json.load(j)

items  = json_dict['data']['morePrints']['items']
print(len(items))

df = pd.DataFrame(items)
print(df.columns)
```

```python
df.drop(columns=['id', 'slug'], inplace=True)
```

```python
df.category = df['category'].apply(lambda cat: cat['path'][0]['name'])

df
```

```python
def process_image_series(row, size: str ='small'):
    """ x
    image url coming in filePaht is like:
        - https://media.printables.com/media/prints/871465/images/6688423_1c4f9d5c-eec5-44d6-ac87-96369d067546_84b87e18-32af-4a96-bc89-b089f5f81fb0/warning-cover.png
    and this url only downloads the image.
    
    We want url to show image on browser, in such format:
        - https://media.printables.com/media/prints/871465/images/6688423_1c4f9d5c-eec5-44d6-ac87-96369d067546_84b87e18-32af-4a96-bc89-b089f5f81fb0/thumbs/inside/320x240/png/warning-cover.webp
    OR full size:
        - https://media.printables.com/media/prints/871465/images/6688423_1c4f9d5c-eec5-44d6-ac87-96369d067546_84b87e18-32af-4a96-bc89-b089f5f81fb0/thumbs/inside/1600x1200/png/warning-cover.webp
    
    Apply this method to a series of raw img data:
        >> df.image.apply(lambda img: process_image_series(img))
    
    - Param (size: str): Can either be full or small which determines size of image, default = 'small'
    """
    raw_url = row['filePath']
    last_slash_index = raw_url.rfind('/')
    
    if size in ['large', 'full']:
        final_url = "https://media.printables.com/" + raw_url[:last_slash_index] + "/thumbs/inside/1600x1200/png" + raw_url[last_slash_index:-4] + ".webp" 
    else:
        final_url = "https://media.printables.com/" + raw_url[:last_slash_index] + "/thumbs/inside/320x240/png" + raw_url[last_slash_index:-4] + ".webp" 
        
    return final_url
```

```python
df.image = df.image.apply(lambda img: process_image_series(img))
df
```

```python
def process_raitng(row):
    if row == '0E-10':
        rating = 0
    else:
        rating = round(float(row), 1)
    return str(rating)
```

```python
df.rating = df.ratingAvg.apply(lambda row: process_raitng(row))
df
```
